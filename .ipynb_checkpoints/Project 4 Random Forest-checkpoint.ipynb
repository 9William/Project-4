```python
#Membagi kolom menjadi kolom numerik dan kolom kategori
var_kategori = [var for var in data.columns if data[var].dtype =='category' and var!='Target']
var_numerik = [var for var in data.columns if data[var].dtype !='category' and var !='Target']

print(var_kategori)
print(var_numerik)
```

```python
#Secara visual
# sns.pairplot(data)
```

#### Pipeine Preprocessing

```python
# Membuat Pipeline untuk preprocessing
preprocessor_numerik = Pipeline([
    ('imputasi', SimpleImputer(strategy='median')),
    ('scaling', StandardScaler())
])

preprocessor_kategori = Pipeline([
    ('imputasi', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('encoding', OneHotEncoder(drop='first', handle_unknown='ignore'))         
])

# Menggabungkan kedua pipeline di atas
preprocessor = ColumnTransformer([
    ('preprocessing numerik', preprocessor_numerik, var_numerik),
    ('preprocessing kategori', preprocessor_kategori, var_kategori)
])
```

#### Fit and transform X_training

```python
# Fit and transform the training data
data_preprocessed = preprocessor.fit_transform(data)

# Extract the column names for one-hot encoded features
onehot_feature_names = preprocessor.named_transformers_['preprocessing kategori'].named_steps['encoding'].get_feature_names_out(var_kategori)

# Combine all feature names
feature_names = var_numerik + list(onehot_feature_names)

# Convert the preprocessed data to a DataFrame
data_preprocessed_df = pd.DataFrame(data_preprocessed, columns=feature_names)

# Display the first few rows of the DataFrame
print(data_preprocessed_df.head())
```

```python
data_baru = pd.DataFrame(data_preprocessed_df, columns=feature_names)
data_baru
```

#### Splitting Data

```python
X = data_baru
y = data['Target']
```

```python
#Melihat Data independen
X.head()
```

```python
#Melihat data dependen 
y.head()
```

#### Mengatasi Dataset imblance 

```python
ros = RandomUnderSampler(
    sampling_strategy='auto', 
    random_state=0,
    replacement=True)

X_res, y_res = ros.fit_resample(X, y)
```

```python
#Hasil proporsi yang baru
print(y_res.value_counts()/len(y_res))
```

#### Data sudah memiliki data yang balance

```python
#Dimensi data yang baru
print(X_res.shape)
print(y_res.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_res,
    y_res,
    test_size=0.2,
    random_state= 0)

#Dimensi
print((X_train.shape), (X_test.shape))
```

```python
#reset index X
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
```

```python
#reset index y
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

#### Feature Selection dengan Metrix ANOVA 

```python
from sklearn.feature_selection import f_classif, f_regression
from sklearn.feature_selection import SelectKBest, SelectPercentile
```

```python
X_train.shape, X_test.shape
```

```python
# Reset indeks dengan looping
for i in [X_train, X_test, y_train, y_test]:
    i.reset_index(drop=True, inplace=True)
```

```python
# Kita hitung nilai f-score dan p-value untuk setiap feature
skor = f_classif(X_train.fillna(0), y_train)

# Melihat skor (terdiri dari f-score dan p-value)
skor
```

```python
skor = pd.Series(skor[1])
skor.index = X_train.columns
skor.sort_values(ascending=True).plot.bar(figsize=(20, 6));
```

semakin kecil nilai p-value, maka semakin bagus feature tersebut untuk dipertahankan.

```python
skor.sort_values(ascending=True)[:12]
```

```python
# Memilih top 10
top_12 = SelectKBest(f_classif, k=12).fit(X_train, y_train)

# Melihat hasilnya
X_train.columns[top_12.get_support()]
```

```python
# buang features yang tidak penting
X_train = top_12.transform(X_train)

# Melihat dimensi X_train
X_train.shape
```

```python
# Melihat X_train
X_train = pd.DataFrame(X_train, columns=top_12.get_feature_names_out())
X_train.head()
```
